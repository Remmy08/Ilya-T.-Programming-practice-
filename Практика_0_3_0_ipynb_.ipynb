# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Ввод:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
list =input().split()
sum = 0
n = 0
max = 0
for i in list:
  sum += int(i)
  n += 1
  if int(i) > int(max):
    max = i
print('Сумма чисел:', sum)
print('Среднее арифм.:', sum/n)
print('Максимальное значение:', max)
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random

KNB = ['Камень', 'Ножницы', 'Бумага']
round = 1
player_score = 0
computer_score = 0
# while (player_score != 3) or (computer_score != 3):
while player_score < 3 and computer_score < 3:

  print('----- Этап', round,'-----')
  player = input('Сделайте выбор: Камень, Ножницы, Бумага. ')
  computer = random.choice(KNB)
  print('Компьютер:', computer)
  print('Игрок:', player)
  if (player == 'Камень' and computer ==  'Ножницы') or (player == 'Ножницы' and computer ==  'Бумага') or (player == 'Бумага' and computer ==  'Камень'):
    print('------------------')
    print('Победа: Игрок')
    player_score += 1
  elif computer == player:
    print('------------------')
    print('Ничья')
  else:
    print('------------------')
    print('Победа: Компьютер')
    computer_score += 1
  round +=1
  print('Счет: Игрок -', player_score, 'Компьютер -', computer_score)

if player_score == 3:
  print('-------------------------------')
  print('Игра окончена! Победил - Игрок!')
  print('-------------------------------')
else:
  print('-----------------------------------')
  print('Игра окончена! Победил - Компьютер!')
  print('-----------------------------------')
````
